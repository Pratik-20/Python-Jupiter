```python
import scipy.ndimage
```

```python
img = "Q.jpeg"
```

```python
def grayscale(rgb):
    return np.dot(rgb[...,:3], [0.400, 0.550,0.100])
```

```python
def dodge(front,black):
    result = front *255/(255-black) 
    result[result>255]=255
    result[black==255]=255
    return result.astype('uint8')
```

```python
s= imageio.imread(img) 
g= grayscale(s) 
i=255-g
```

```python
b= scipy.ndimage.filters.gaussian_filter(i, sigma=10) 
r=dodge(b,g) 
cv2.imwrite('00.png', r)
```
